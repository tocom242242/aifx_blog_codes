```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
dataset = load_boston()
df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```
plt.rcParams['font.size'] = 10
fig, ax = plt.subplots(figsize=(10,10))

df.hist(ax=ax)
```
