```
from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use("bmh")
```

```
lams = [3, 5, 7, 10, 20]
colors = ["k", "b", "g", "r", "y"]
fig, ax = plt.subplots(1, 1, figsize=(5,5))
x = np.arange(30)
for lam,c in zip(lams, colors):
    ax.plot(x, poisson.pmf(x,lam), c, lw=4, alpha=0.9, label=f'$\lambda$ = {lam}')
    ax.legend()
fig.subplots_adjust(wspace=0.5, hspace=0.5)
```
